```python
# Filling missing values

# gen=df['Gender'].mode()[0]
# df['Gender'].fillna(gen,inplace=True)

df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])

df['Married']=df['Married'].fillna(df['Married'].mode()[0])

df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])

df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])

df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())

df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())

df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mean())
df.isna().sum()
```

```python
# Encoding

from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()

df['Gender']=lb.fit_transform(df['Gender'])

df['Married']=lb.fit_transform(df['Married'])

df['Dependents']=lb.fit_transform(df['Dependents'])

df['Education']=lb.fit_transform(df['Education'])

df['Self_Employed']=lb.fit_transform(df['Self_Employed'])

df['Property_Area']=lb.fit_transform(df['Property_Area'])

df.dtypes
```

```python
df.drop(['Loan_ID'],axis=1,inplace=True)
df
```

```python
# Splitting Input and Output data

x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x
```

```python
y
```

```python
# Splitting Training and Testing Data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
# Standard Scaler Normalization

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# Model Creation

from sklearn.naive_bayes import BernoulliNB # if the data contains negative values it will not support MultinomialNB, So we use BernoulliNB
model=BernoulliNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Performance Evaluation

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
# Accuracy Score

score=accuracy_score(y_test,y_pred)
score
```

```python
# Classification Report

report=classification_report(y_test,y_pred)
print(report)
```

```python
# Graphical Representation

from sklearn.metrics import ConfusionMatrixDisplay
labels=['Y','N']
cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
cmd.plot()
```
